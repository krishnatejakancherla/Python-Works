```python
#Part - 3 - Seaborn Axis Grids (Pair Grid, Joint Grid, Facet Grid)
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.info()
```

```python
iris.head()
```

```python
tips['day'].nunique(), tips['time'].nunique(), tips['smoker'].nunique()
```

```python
#FacetGrid
sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g = g.map(plt.hist, 'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker', hue = 'sex')
g = g.map(plt.scatter, 'total_bill', 'tip').add_legend()
```

```python
#PairGid - is a sub plot grid
sns.PairGrid(iris)
```

```python
g =sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g =sns.PairGrid(iris, hue = 'species')
g.map_diag(plt.hist) # Histograms in Diagonal
g.map_offdiag(plt.scatter) #Scatters in all other placees
g.add_legend()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot) #dist at diagonal
g.map_upper(plt.scatter) # Scatter at upper
g.map_lower(sns.kdeplot) #kde at lower
```

```python
sns.pairplot(iris, hue = 'species', palette = 'rainbow')
```

```python
#Joint Grid
sns.jointplot(x ='total_bill', y = 'tip', data = tips)
```

```python
g = sns.JointGrid(x ='total_bill', y = 'tip', data = tips) #Initialize
g.plot(sns.regplot, sns.distplot) # put data
```
