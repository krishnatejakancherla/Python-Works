```python
# *******************************Python For Data Visualisation using Matplotlib ****************************************
#Matplotlib Essentials 
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Part - 1 - Basic Plotting and Object oriented Approach
import numpy as np
```

```python
x = np.arange(11)
y = x ** 2
```

```python
x, y
```

```python
# Basic plot - Line plot
plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Figure X and X**2')
#plt.show() # prints the data plot
```

```python
#Multiple plot on same canverse
plt.subplot(1, 3, 1)
plt.plot(x,y, 'r--') # color and dotted lines

plt.subplot(1, 3, 2)
plt.plot(y,x, 'g--')

plt.subplot(1, 3, 3)
plt.plot(x,y, 'g*--') # Star and dotted lines
plt.plot(y,x, 'r*-')

plt.tight_layout()
```

```python
# Part - 2 - Basic Plotting and Object oriented Approach
#Object Oriented approach -  recomended for more control and customisation of plots
fig1 = plt.figure()
axes = fig1.add_axes([0.1,0.1,0.9,0.9]) # left, bottom, width, height
axes.plot(x,y) # Plot data
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Graph')
```

```python
fig2 = plt.figure()
axes_main = fig2.add_axes([0.1,0.1,0.9,0.9])
axes_inset = fig2.add_axes([0.3,0.5,0.3,0.4])
axes_main.plot(x,y,'b')
axes_inset.plot(y,x,'r')

axes_main.set_xlabel('Main X')
axes_main.set_ylabel('Main Y')
axes_main.set_title('MAIN')

axes_inset.set_xlabel('Inset X')
axes_inset.set_ylabel('Inset Y')
axes_inset.set_title('INSET')

```

```python
# Create afigure with set of sub plots
fig , axes = plt.subplots(nrows = 2, ncols = 2, sharex = True, sharey = True)
plt.tight_layout()
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig1, axes1 = plt.subplots(nrows=2, ncols=1)
fig2, axes2 = plt.subplots(nrows=1, ncols=2)
```

```python
type(axes1), type(axes2)
```

```python
for ax in axes1:
    ax.plot(x,y)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('XY Title')
fig1.tight_layout() # to overcome overlapping
fig1
```

```python
axes2[0].plot(x,y,'b')
axes2[0].set_xlabel('A')
axes2[0].set_ylabel('B')
axes2[0].set_title('first')


axes2[1].plot(x,y,'g')
axes2[1].set_xlabel('A')
axes2[1].set_ylabel('B')
axes2[1].set_title('second')

fig2.tight_layout()
fig2
```

```python
# Part - 3 - Basic Plotting and Object oriented Approach
#Figure Size - Pixels per inch - DPI
fig = plt.figure(figsize = (10,5), dpi = 100)
```

```python
fig, axes = plt.subplots(figsize = (10,5))
axes.plot(x,y,'y')
```

```python
#SAve figure Fig
fig.savefig('figname.png', dpi = 100)
```

```python
#Legends on Plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='x**2')
axes.plot(x,x**3,label='x**3')
axes.legend(loc = 2)  # displays legends - 1= upper right corner, 2 = upper left corner, 3= lower right corner, 4 = lower left corner
```

```python
fig, axes = plt.subplots()
axes.plot(x, x+1, 'r.-')
axes.plot(x, x+2, 'g--')
```

```python
fig, axes = plt.subplots()
axes.plot(x, x+1, color = 'blue', alpha = 0.8)
axes.plot(x, x+2, color = 'green')
axes.plot(x, x+3, color = 'red')
```

```python
#Line width, style , marker style and others
fig, axes = plt.subplots(figsize = (10,8))
axes.plot(x,y, color = 'purple', lw = 4, ls = '-', marker = 's', markersize = 18, 
          markerfacecolor = 'yellow', markeredgewidth = 4, markeredgecolor = 'red')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (10,3))
axes[0].plot(x,y)
axes[0].set_title('default')


axes[1].plot(x,y)
axes[1].axis('tight')
axes[1].set_title('tight')

axes[2].plot(x,y)
axes[2].set_title('custom')
axes[2].set_xlim([1,4])
axes[2].set_ylim([0,50])
```
