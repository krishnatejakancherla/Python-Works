```python
#*********************Python for interactive & geographical plotting using Plotly and Cufflinks***************
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
from plotly import __version__
print(__version__)
```

```python
from plotly.offline import download_plotlyjs
from plotly.offline import init_notebook_mode
from plotly.offline import plot, iplot
```

```python
import cufflinks as cf
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df1 = pd.DataFrame(np.random.randn(100,4), columns = ['A','B','C','D'])
df1.head()
```

```python
dic = {'fruit':['apple','orange','banana', 'grape'], 'count':[32,24,33,54]}
```

```python
df2 = pd.DataFrame(dic)
df2.head()
```

```python
df1.plot()
```

```python
df1.iplot() #interactive plot
```

```python
df1.iplot(kind = 'scatter', x = 'A', y = 'B', mode = 'markers', size = 10)
```

```python
df2.iplot(kind = 'bar', x ='fruit', y = 'count')
```

```python
df1.iplot(kind = 'box')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5], 'y':[10,20,30,20,10], 'z':[5,4,3,2,1]})
df3.head()
```

```python
df3.iplot(kind = 'surface')
```

```python
df1['A'].iplot(kind = 'hist', bins = 50)
```

```python
df1[['A', 'B']].iplot(kind = 'spread')
```

```python
df1.iplot(kind = 'bubble', x='A', y = 'B', size = 'C')
```

```python
df1.scatter_matrix()
```
