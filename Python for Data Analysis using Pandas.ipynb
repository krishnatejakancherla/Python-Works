```python
# Part - 5 - Data Wrangling - Combining. Joining , Merging
d1 = {'key':['a','b','c','d','e'], 'A1': range(5), 'B1': range(5,10)}
d2 = {'key':['a','b','c'], 'A2': range(3), 'B2': range(3,6)}
```

```python
d1
```

```python
d2
```

```python
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
 # merge the Data Frame -  inner , outer, left, right
pd.merge(df1, df2, how = 'inner', on = 'key')
```

```python
pd.merge(df1, df2, how = 'outer', on = 'key')
```

```python
pd.merge(df1, df2, how = 'left', on = 'key')
```

```python
pd.merge(df1, df2, how = 'right', on = 'key')
```

```python
left = pd.DataFrame({'key1':['a','a','b','c'],
                     'key2':['a','b','a','b'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3'],
                    })

right = pd.DataFrame({'key1':['a','b','b','c'],
                     'key2':['a','b','a','a'],
                     'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3'],
                    })
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how = 'inner', on = ['key1','key2']) # gives only which are present in both
```

```python
pd.merge(left, right, how = 'outer', on = ['key1','key2']) #union op, it gets all 
```

```python
pd.merge(left, right, how = 'left', on = ['key1','key2']) 
```

```python
pd.merge(left, right, how = 'right', on = ['key1','key2']) 
```

```python
# Part - 6 - Group by - SPlit, apply and combine
data = {'Store': ['Walmart', 'Walmart', 'Costco', 'Costco', 'Target', 'Target'],
        'Customer':['KT', 'Abhigna', 'Prashant', 'Sravan', 'Mahesh', 'Indu'],
        'Sales': [200,150,100,300,250,90]}

df = pd.DataFrame(data)
df
```

```python
#Grup the data based on Stores
by_store = df.groupby('Store')
by_store
```

```python
by_store.mean()
```

```python
df.groupby('Store').sum().loc['Target']
```

```python
by_store.min()
```
